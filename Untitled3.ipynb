```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10

# تحميل الداتا
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# preprocessing
x_train = x_train / 255.0
x_test = x_test / 255.0

from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# ناخد جزء صغير عشان السرعة
x_train_small = x_train[:2000]
y_train_small = y_train[:2000]

# ================= MODEL =================
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Input, Lambda, GlobalAveragePooling2D
from tensorflow.keras.models import Model

input_layer = Input(shape=(32, 32, 3))

# resize
resized = Lambda(lambda image: tf.image.resize(image, (224, 224)))(input_layer)

# base model
base = MobileNetV2(include_top=False, weights='imagenet')

x = base(resized)

# أهم تعديل 🔥
x = GlobalAveragePooling2D()(x)

model = Model(inputs=input_layer, outputs=x)

print("Model ready ✅")

# ================= FEATURE EXTRACTION =================
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

x_train_prep = preprocess_input(x_train_small)

features_train = model.predict(x_train_prep)

print("Features shape:", features_train.shape)
# المفروض (2000, 1280)

# ================= SVM =================
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

y_train_labels = np.argmax(y_train_small, axis=1)

svm = SVC(kernel='linear')
svm.fit(features_train, y_train_labels)

y_pred = svm.predict(features_train)

acc = accuracy_score(y_train_labels, y_pred)
print("Accuracy:", acc)
# ناخد test صغير برضه
x_test_small = x_test[:1000]
y_test_small = y_test[:1000]

# preprocess
x_test_prep = preprocess_input(x_test_small)

# features
features_test = model.predict(x_test_prep)

# labels
y_test_labels = np.argmax(y_test_small, axis=1)

# prediction
y_pred_test = svm.predict(features_test)

# accuracy
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

acc_test = accuracy_score(y_test_labels, y_pred_test)
print("Test Accuracy:", acc_test)
cm = confusion_matrix(y_test_labels, y_pred_test)
print("Confusion Matrix:\n", cm)
```
